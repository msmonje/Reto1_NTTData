#### Vamos a asumir que la tasa de crecimiento de poblaci√≥n para 2002 de los estados de Alabama y South Carolina son dos n√∫meros aleatorios entre 0 y 0.1. Asignaremos a Alabama el ratio de crecimiento menor y a South Carolina el ratio de crecimiento mayor. Vamos a asumir tambi√©n la poblaci√≥n de 2001 de ambos estados descrita en los ejercicios anteriores.

#### Con esta informaci√≥n:

#### a. ¬øCu√°ntos a√±os tardar√° el estado de South Carolina en alcanzar en poblaci√≥n a Alabama?

#### c. Si al planteamiento anterior le a√±adimos la tasa de fallecidos de cada estado. ¬øCu√°ntos a√±os tardar√° South Carolina en alcanzar en poblaci√≥n a Alabama? Vamos a asumir la tasa de fallecidos y la poblaci√≥n correspondientes a 2001.

```python
# a. ¬øCu√°ntos a√±os tardar√° el estado de South Carolina en alcanzar en poblaci√≥n a Alabama? 
import random # importando la librer√≠a random

# Pariables iniciales y las tasas de crecimiento aleatorias
# Poblaci√≥n de Alabama y South Carolina en 2001
pob_alabama_2001 = estados[0]['poblacion_2001'] 
pob_south_carolina_2001 = estados[3]['poblacion_2001']

# Tasas de crecimiento aleatorias entre 0 y 0.1
crecimiento_alabama = random.uniform(0, 0.1)
crecimiento_south_carolina = random.uniform(crecimiento_alabama, 0.1)

while crecimiento_alabama == crecimiento_south_carolina:
    crecimiento_alabama = random.uniform(0, 0.1)
    crecimiento_south_carolina = random.uniform(crecimiento_alabama, 0.1)

print('Tasa de crecimiento de Alabama:', crecimiento_alabama)
print('Tasa de crecimiento de South Carolina:', crecimiento_south_carolina)

# ¬øCu√°ntos a√±os tardar√° South Carolina en alcanzar a Alabama en poblaci√≥n?
anios = 0
while pob_south_carolina_2001 <= pob_alabama_2001:
    pob_alabama_2001 *= 1 + crecimiento_alabama
    pob_south_carolina_2001 *= 1 + crecimiento_south_carolina
    anios += 1

print(f'South Carolina superar√° a Alabama en poblaci√≥n en {anios} a√±os')

# b. ¬øEn qu√© a√±o ocurrir√° esto?
anios_para_superar = 2001 + anios
print(f'Esto ocurrir√° en el a√±o {anios_para_superar}')
```

```python
# c. Si al planteamiento anterior le a√±adimos la tasa de fallecidos de cada estado: 
# ¬øCu√°ntos a√±os tardar√° South Carolina en alcanzar en poblaci√≥n a Alabama? 
# Vamos a asumir la tasa de fallecidos y la poblaci√≥n correspondientes a 2001.

# Poblaci√≥n y muertes en Alabama y South Carolina en 2001
pob_alabama_2001 = estados[0]['poblacion_2001'] 
pob_south_carolina_2001 = estados[3]['poblacion_2001']
muertes_alabama_2001 = estados[0]['muertes_2001']
muertes_south_carolina_2001 = estados[3]['muertes_2001']

# Tasas de crecimiento aleatorias entre 0 y 0.1
crecimiento_alabama = random.uniform(0, 0.1)
crecimiento_south_carolina = random.uniform(crecimiento_alabama, 0.1)

# Tasas de mortalidad que se agregan al mismo c√°lculo
mortalidad_alabama = muertes_alabama_2001 / pob_alabama_2001 # tasa de mortalidad de Alabama en 2001
mortalidad_south_carolina = muertes_south_carolina_2001 / pob_south_carolina_2001 # tasa de mortalidad de South Carolina en 2001

# Inicializando anios de diferencia en 0    
anios = 0
while pob_south_carolina_2001 <= pob_alabama_2001:
    pob_alabama_2001 *= 1 + crecimiento_alabama - mortalidad_alabama
    pob_south_carolina_2001 *= 1 + crecimiento_south_carolina - mortalidad_south_carolina
    anios += 1

print(f'South Carolina superar√° a Alabama en poblaci√≥n (teniendo en cuenta la mortalidad) en {anios} a√±os')
```

#### `Ejercicio 3`

#### Cread una funci√≥n que genere una proyecci√≥n para el a√±o 2002, utilizando como ratio la comparativa entre los a√±os 2000 y 2001 de tal forma:

Poblaci√≥n A√±o 2002 = Poblaci√≥n A√±o 2001 / Poblaci√≥n A√±o 2000 x Poblaci√≥n a√±o 2001.
Tras un estudio demogr√°fico se ha determinado que el n√∫mero de habitantes de cierta 
poblaci√≥n, en los pr√≥ximos a√±os, vendr√° dado por la funci√≥n: P(t) = 14500t + 7000/2t + 1 
donde t son los a√±os transcurridos entre la fundaci√≥n del Estado y 1900.
En la misma funci√≥n, devolved la poblaci√≥n que tendr√≠a cada Estado bajo la estimaci√≥n que 
devuelve ese estudio demogr√°fico.

A√±adid los resultados a cada diccionario de la lista

```python
# Funci√≥n de proyecci√≥n de poblaci√≥n
def proyeccion_poblacion(poblacion_2000, poblacion_2001):
    poblacion_2002 = poblacion_2001 / poblacion_2000 * poblacion_2001
    return round(poblacion_2002,1)

# Funci√≥n sobre el estudio demogr√°fico
def poblacion_estado(anios_transcurridos):
    poblacion = 14500 * anios_transcurridos + 7000 / (2 * anios_transcurridos + 1)
    return poblacion

lista_estados = [estados[0]['nombre'], estados[1]['nombre'], estados[2]['nombre'], estados[3]['nombre']]
poblacion_2000 = [estados[0]['poblacion_2000'], estados[1]['poblacion_2000'], estados[2]['poblacion_2000'], estados[3]['poblacion_2000']]
poblacion_2001 = [estados[0]['poblacion_2001'], estados[1]['poblacion_2001'], estados[2]['poblacion_2001'], estados[3]['poblacion_2001']]

fecha_fundacion_Alabama = 1819
fecha_fundacion_Florida = 1845
fecha_fundacion_Georgia = 1733
fecha_fundacion_South_Carolina = 1776

anios_transcurridos = [1900 - fecha_fundacion_Alabama, 1900 - fecha_fundacion_Florida, 1900 - fecha_fundacion_Georgia, 1900 - fecha_fundacion_South_Carolina]

resultados = []
for i in range(len(lista_estados)):
    diccionario = {}
    diccionario["Estado"] = lista_estados[i]
    diccionario["Poblaci√≥n 2002"] = proyeccion_poblacion(poblacion_2000[i], poblacion_2001[i])
    diccionario["Poblaci√≥n Estado"] = round(poblacion_estado(anios_transcurridos[i]), 1)
    resultados.append(diccionario) 

print(resultados)
    
```
